```python
important_columns = df[[
     "OverallQual",
    "GrLivArea",
    "GarageCars",
    "GarageArea",
    "TotalBsmtSF",
    "FullBath",
    "YearBuilt",
    "1stFlrSF",
    "TotRmsAbvGrd",
    "SalePrice"
]]
```

```python
important_columns.info()
important_columns.describe()
```

```python


important_columns.loc[:,"GrLivArea"] = important_columns["GrLivArea"].clip(upper=3700)
important_columns.loc[:,"GarageCars"] = important_columns["GarageCars"].clip(upper=3)
important_columns.loc[:,"GarageArea"] = important_columns["GarageArea"].clip(upper=1200,lower=100)
important_columns.loc[:,"TotalBsmtSF"] = important_columns["TotalBsmtSF"].clip(upper=3000)
important_columns.loc[:,"1stFlrSF"] = important_columns["1stFlrSF"].clip(upper=3000)
important_columns.loc[:,"TotRmsAbvGrd"] = important_columns["TotRmsAbvGrd"].clip(upper=12)

```

```python
from turtle import color
import matplotlib.pyplot as plt
features = [ "OverallQual",
    "GrLivArea",
    "GarageCars",
    "GarageArea",
    "TotalBsmtSF",
    "FullBath",
    "YearBuilt",
    "1stFlrSF",
    "TotRmsAbvGrd",
    "SalePrice"]
for feature in features:
    plt.figure(figsize=(8,6))
    plt.scatter(important_columns[feature],important_columns["SalePrice"],color="blue",alpha=0.5)
    plt.xlabel(feature)
    plt.ylabel("Sales Price")
    plt.title(f"{important_columns[feature]} vs salesprice")
    plt.show()
```

```python
import seaborn as sns

corr = important_columns.corr()
plt.figure(figsize=(7,5))
sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.show()
```

```python
X = important_columns.iloc[:,:-1]
y = important_columns.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=45)
```

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
lr.fit(X_train,y_train)
```

```python
import numpy as np
y_pred = lr.predict(X_test)
from sklearn.metrics import r2_score,mean_squared_error
r2 = r2_score(y_test,y_pred)
mse = mean_squared_error(y_test,y_pred)
print("r2 score = ",r2)
print("Rmse = ",np.sqrt(mse))
```

```python
print("Accuracy = ",r2*100)
```

```python

plt.figure(figsize=(6,6))
plt.scatter(range(len(y_test)), y_test, color='blue', label='Actual SalePrice', alpha=0.6)
plt.scatter(range(len(y_pred)), y_pred, color='green', label='Predicted SalePrice', alpha=0.6)

plt.xlabel("Sample Index")
plt.ylabel("SalePrice")
plt.title("Actual vs Predicted SalePrice")
plt.legend()
plt.tight_layout()
plt.savefig("actual_vs_predicted.png", bbox_inches='tight')
plt.show()
plt.close()
```
